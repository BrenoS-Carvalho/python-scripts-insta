```python
import pandas as pd
tabela = pd.read_excel('vendas.xlsx')
display(tabela)
#tabela original
```

```python
#vou criar uma coluna com os totais de cada linha
tabela["Total"] = tabela["PrecoUnitario"]*tabela["Qtd. Vendida"]
display(tabela)
#percebe que o pandas criou a coluna Total
```

```python
#vamos ver o total das vendas
total_vendido = tabela["Total"].sum()
print(f'O total vendido foi de R$ {total_vendido:.2f}')
```
