```python
import matplotlib.pyplot as plt
x = ['Maio', 'Junho', 'Julho', 'Agosto', 'Setembro']
y = [3500, 2800, 2900, 5400, 4300]
plt.plot(x, y, 'g')
plt.plot(x, y, "r o")
plt.ylim(2500, 6000)
plt.title('Produção anual (Maio a Setembro)')
plt.xlabel('Meses')
plt.ylabel('Valores em R$')
plt.show()
```
